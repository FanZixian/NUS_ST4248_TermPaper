# Data Cleaning and Manipulation

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
path = "./Data/"
```

```python
# Reading csv files from local repository
def read(pathname):
    df = pd.read_csv(pathname, header=[0])
    return df
```

```python
HDBDataset = read(path+"Source/HDB Rental Statistics.csv")
```

```python
ProcessedHDB = pd.DataFrame(columns = ["Rental_Price", "Time", "District", "RoomType"])
```

```python
for index, row in HDBDataset.iterrows():
    year = row['Year']
    quarter = row['Quarter']
    
    time = f"Q{quarter} {year}"
    
    district = row['District']
    
    # Create a new row for each room type with the corresponding rental price, time, district, and room type
    ProcessedHDB = ProcessedHDB.append({"Rental_Price": row['Room3'], "Time": time, "District": district, "RoomType": "Room3"}, ignore_index=True)
    ProcessedHDB = ProcessedHDB.append({"Rental_Price": row['Room4'], "Time": time, "District": district, "RoomType": "Room4"}, ignore_index=True)
    ProcessedHDB = ProcessedHDB.append({"Rental_Price": row['Room5'], "Time": time, "District": district, "RoomType": "Room5"}, ignore_index=True)
    ProcessedHDB = ProcessedHDB.append({"Rental_Price": row['Executive'], "Time": time, "District": district, "RoomType": "Executive"}, ignore_index=True)
```

```python
ProcessedHDB = ProcessedHDB.dropna()
```

```python
ProcessedHDB.to_csv(path+"HDB_Parametric.csv")
```

```python
IR = read(path+"Source/InterestRate.csv")
```

```python
IR["Time"] = pd.to_datetime(IR["Time"], format='%Y %b ', errors='coerce')
```

```python
IR.set_index('Time', inplace=True)
```

```python
quarterly_mean = IR.resample('Q').mean()
```

```python
quarterly_mean.to_csv(path+"Source/IRQuarterly.csv")
```

```python
Macro = read(path+"Source/Macro.csv")
```

```python
Macro["Time"] = Macro["Time"].apply(lambda x: 'Q' + x.split()[1][0] + ' ' + x.split()[0])
```

```python
merged_df = pd.merge(ProcessedHDB, Macro, on='Time', how='left')
```

```python
merged_df.to_csv(path+"TotalData.csv")
```
